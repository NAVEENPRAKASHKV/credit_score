```python
# converting Credit_History_Age to number of days

def days_conversion(val):
    if pd.isna(val):
        return np.nan
    try:
        year,month=map(int,val.split()[0::3])
        days=(year*365)+(month*30)
        return days
    except Exception as e:
        print(e)
df['Credit_History_Age']=df['Credit_History_Age'].apply(days_conversion)
```

# Month

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoder.classes_ =['January' ,'February', 'March', 'April' ,'May' ,'June' ,'July', 'August']
df['Month']=encoder.fit_transform(df['Month'])
```

# Occupation

```python
mapping={'Scientist':1, '_______':np.nan, 'Teacher':2, 'Engineer':3, 'Entrepreneur':4, 'Developer':5,
 'Lawyer':6, 'Media_Manager':7, 'Doctor':8, 'Journalist':9, 'Manager':10, 'Accountant':11,'Musician':12, 'Mechanic':13, 'Writer':14, 'Architect':15}
df['Occupation']=df['Occupation'].map(mapping)
```

```python
df['Occupation'].unique()
```

# Credit_Mix

```python
#all other category ie  _ total 20195 are converted to -1 we need to handil this value

mapped=['Bad','Standard','Good']

from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[mapped],handle_unknown='use_encoded_value', unknown_value=-1)

input_data=df['Credit_Mix'].values.reshape(-1,1)
df['Credit_Mix']=encoder.fit_transform(input_data)

```

# Payment_of_Min_Amount

```python
mapped={ 'No':1, 'NM':np.nan, 'Yes':2}
df['Payment_of_Min_Amount']=df['Payment_of_Min_Amount'].map(mapped)
```

# Payment_Behaviour

```python
mapped={'High_spent_Small_value_payments':1 ,'Low_spent_Large_value_payments':2,
 'Low_spent_Medium_value_payments':3, 'Low_spent_Small_value_payments':4,
 'High_spent_Medium_value_payments':5, '!@9#%8':np.nan,
 'High_spent_Large_value_payments':6}

df['Payment_Behaviour']=df['Payment_Behaviour'].map(mapped)
```
